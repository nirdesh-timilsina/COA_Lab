# Objective : to design N- bit adder for unsigned integer binary numbers

```python
#first input 
print ("Enter n bit binary strings: ")
A = input("A: ")
if any (el not in '01' for el in A):
   raise ValueError("Only enter binary number.")
A
```

```python
B = input("B:")
if any (el not in '01' for el in B):
    raise ValueError ("Only enter binary number.")
B
```

```python
if len(A) != len(B):
    raise ValueError("both inputs must be of same bit length.")
```

```python
length = len(A)
```

```python
def fullAdder( a,  b,  c):
    s = a^b^c
    Carry = a*b + (a^b)*c
    return s,Carry
```

```python
s,c = 0,0
sum_ans =""
for i in range (length -1 ,-1,-1):
    a,b = int(A[i]),int(B[i])
    s,c = fullAdder(a,b,c)
    sum_ans = str(s)+sum_ans

print (f"A: {A},B: {B},sum :{sum_ans}, carry : {c}")
```

Conclusion:
An n-bit unsigned binary adder was successfully implemented by cascading full adder logic in software. The program dynamically adapts to the input bit-width and correctly propagates carry across all bit positions. The design accurately models a ripple-carry adder and produces correct sum and carry outputs for arbitrary-length binary inputs.
